```python
import pandas as pd
import numpy as np
from matplotlib import pylab as plt
#import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Complex network Restaurant.csv")
data=data.drop(["Horodateur","Score"], axis=1)
#data=np.array(data
data.head()
```

```python
import networkx as nx
G = nx.Graph()
```

```python
for row in data.iterrows():
       G.add_edges_from([(row[1][0],row[1][1]), (row[1][0],row[1][2])])
```

```python
#initialze Figure
plt.figure(num=None, figsize=(20, 20), dpi=100)
plt.axis('off')
fig = plt.figure(1)
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)
cut = 0.80
# xmax = cut * max(xx for xx, yy in pos.values())
# ymax = cut * max(yy for xx, yy in pos.values())
# plt.xlim(0, xmax)
# plt.ylim(0, ymax)
plt.savefig("resto.png")
#pylab.close()
del fig
```

```python
def density(G):
    n=nx.number_of_nodes(G)
    Lmax=(n*(n-1))/2
    return nx.number_of_edges(G)/Lmax
```

```python
def average_clustering(G):
    return sum(nx.clustering(G).values())/nx.number_of_nodes(G)
```
